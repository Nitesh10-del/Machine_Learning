```python
import seaborn as sns
sns.displot(df1['Student ID'],kind='kde')
```

```python
df1['Student ID']=df1['Student ID'].fillna(df1['Student ID'].mean)
df1['Student ID'].isnull().sum()
```

```python
sns.displot(df1['Class'],kind='kde')
```

```python
df1['Class']=df1['Class'].fillna(df1['Class'].mean)
df1['Class'].isnull().sum()
```

```python
sns.displot(df1['Study hrs'],kind='kde')
```

```python
df1['Study hrs']=df1['Study hrs'].fillna(df1['Study hrs'].mean)
df1['Study hrs'].isnull().sum()
```

```python

sns.displot(df1['Social Media usage hrs'],kind='kde')
```

```python
df1['Social Media usage hrs']=df1['Social Media usage hrs'].fillna(df1['Social Media usage hrs'].median)
df1['Social Media usage hrs'].isnull().sum()
```

```python
sns.displot(df1['Percentage'],kind='kde')
```

```python
df1['Percentage']=df1['Percentage'].fillna(df1['Percentage'].median)
df1['Percentage'].isnull().sum()
```
