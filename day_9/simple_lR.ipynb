```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\nk424\OneDrive\Scans\Desktop\training\Sample_Data\Salary_Data.csv")
df.head()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,1].values
```

```python
type(x)
type(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
print("shape of x_train",x_train.shape)
print("shape of x_test",x_test.shape)
print("shape of Y_train",y_train.shape)
print("shape of Y_test",y_test.shape)


```

```python
# fit_intercept	= Whether to calculate intercept
# copy_X	= Copy original data or not
# n_jobs	= Number of CPU cores
# positive	= Force coefficients to be positive
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x_train, y_train)

```

```python
y_pred=lr.predict(x_test)
x_pred=lr.predict(x_train)
```

```python
# for trainoing data
plt.scatter(x_train,y_train,color='g')
plt.plot(x_train,x_pred,color='r')
plt.show()
```

```python
# for testing data
plt.scatter(x_test,y_test,color='g')
plt.plot(x_train,x_pred,color='r')
plt.show()
```

```python
lr.score(x_test,y_test)
```

```python
# calc slope and inte
# rsept
m=lr.coef_
m
```
